```python
from ollama import chat

stream = chat(
    model='llama3.2:1b',
    messages=[{'role': 'user', 'content': 'what are the best landmarks in london'}],
    stream=True,
    options={"temperature": 0.0} # set between 0.0 to 2.0+
)

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
```
